## Linear Regression Model with single input feature

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate random data
np.random.seed(42)
X=2*np.random.rand(100,1)
y = 4+ 3*X + np.random.randn(100,1)

# Plot the data
plt.scatter(X,y,color="blue")
plt.title("Generated Data")
plt.xlabel("Feature(X)")
plt.ylabel("Target (y)")
plt.show()
```

```python
#Initialize weights and bias
w = np.random.randn(1)
b = np.random.randn(1)

# Learning rate and iterations
learning_rate = 0.001
iterations = 1000

print(w,b)
```

```python
# Predict function
def predict(X,w,b):
    return X @ w + b

#Cost function
def compute_cost(y,y_pred):
    return np.mean((y-y_pred)**2)
```

```python
# Gradient Descent
for i in range(iterations):
    #Predictions
    y_pred = predict(X,w,b)

    #Gradients
    dw = -2 * np.mean(X*(y-y_pred))
    db = -2 * np.mean(y-y_pred)

    #Update weights and bias
    w -= learning_rate * dw
    b -= learning_rate * db

    # Print cost every 100 iterations
    if i % 100 == 0:
        cost = compute_cost(y,y_pred)
        print(f"Iteration {i}, Cost: {cost:4f}")
```

```python
plt.scatter(X,y,color="blue",label = "Data")
plt.plot(X,predict(X,w,b),color="red",label="Regression Line")
plt.title("Linear Regression Fit")
plt.xlabel("Feature(X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()
```

```python
#final model parameters
print(f"Final weight: {w[0]:.4f}")
print(f"Final bias:{b[0]:.4f}")
```

## Linear Regression with 5 input features

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate weights for each feature
true_weights = np.array([[3], [2], [1], [0.5], [-1]])

n_samples = 1000
features = 5

# Dataset
X = 2 * np.random.rand(n_samples,features)
y = 4 + X @ true_weights + np.random.randn(n_samples,1)
```

```python
# Linear Regression Model
def predict(X,w,b):
    return X @ w + b

# Cost Calculation Function
def  compute_cost(y,y_pred):
    return np.mean((y-y_pred) ** 2)
```

```python
def gradient_descent(X,y,w = np.random.rand(5,1),b = np.random.randn(1),learning_rate = 0.001, iterations = 1000):

    for i in range (iterations):

        y_pred = predict(X,w,b)
        #Gradients 

        dw = (1/n_samples) * (X.T @ (y_pred - y).reshape(-1,1))
        db = (1/n_samples) * np.sum(y_pred-y)

        w -= learning_rate * dw
        b -= learning_rate * db

        if i%100 == 0:
            cost = compute_cost(y,y_pred)
            print(f"Iteration {i}, Cost: {cost:.4f}")
    return w,b
```

```python
final_weights, final_bias = gradient_descent(X,y)
print(true_weights)
print(final_weights,final_bias)
```
