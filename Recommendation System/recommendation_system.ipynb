## Cleaning Product Description

```python
import re
# Function to remove HTML tags using regex
def remove_html_tags(text):
    if isinstance(text, str):  # Only process strings
        # Remove HTML tags
        text = re.sub(r'<.*?>', '', text)
        # Remove \r and \n and other escape sequences
        text = re.sub(r'\\[rnt]', ' ', text)
        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Use regex to remove the non-alphanumeric characters
        text = re.sub(r'[^\x00-\x7F]+', '', text)
        return text
    else:
        return text  # Return as-is if not a string
```

```python
# Applying to the DataFrame column
df_product['product_description'] = df_product['product_description'].apply(remove_html_tags)
df_product.head()
```

## Creating user-item matrix

```python
# Creating a pivot table for collaborative filtering
user_item_matrix = df_sales_filtered.pivot_table(index='customer_id', columns='product_id', values='quantity', fill_value=None)
```

```python
user_item_matrix.head()
```

## Centering around 0

```python
avg_quantity = user_item_matrix.mean(axis = 1)

user_item_matrix_centered = user_item_matrix.sub(avg_quantity,axis = 0)

user_item_matrix_centered.fillna(0,inplace=True)
```

```python
user_item_matrix_centered.head()
```

## Performing SVD

```python
# Import the required libraries
from scipy.sparse.linalg import svds
from scipy.sparse import csr_matrix

# Convert the Pandas DataFrame to a sparse matrix
user_item_matrix_sparse = csr_matrix(user_item_matrix_centered.values)

# Decompose the matrix
U, sigma, Vt = svds(user_item_matrix_sparse)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
# Dot product of U and sigma
U_sigma = np.dot(U, sigma)

# Dot product of result and Vt
U_sigma_Vt = np.dot(U_sigma, Vt)

# Add back on the row means contained in avg_ratings
uncentered_quantities = U_sigma_Vt + avg_quantity.values.reshape(-1, 1)


# Create DataFrame of the results
calc_pred_quantities_df = pd.DataFrame(uncentered_quantities,
                                    index=user_item_matrix.index,
                                    columns=user_item_matrix.columns
                                   )
```

```python
# Count how many unique customers have purchased each product
product_popularity = df_sales.groupby('product_id')['customer_id'].nunique().sort_values(ascending=False)
product_popularity.head()
```

```python
product_popularity = product_popularity / df_sales['customer_id'].nunique()  # Normalize
product_popularity.tail()
```

```python
product_popularity = product_popularity.to_dict()
```

```python
boundary_index = int(len(product_popularity) * 0.80)
```

```python
boundary_value = list(product_popularity.values())[boundary_index]
```

```python
boundary_value
```

```python
def uncommon_products(customer_id, threshold=boundary_value):
    # Filter purchases made by the customer
    customer_purchases = df_sales[df_sales['customer_id'] == customer_id]['product_id'].unique()

    uncommon = []
    for product in customer_purchases:
      if product_popularity.get(product) < threshold:
        uncommon.append(product)

    # # Identify uncommon products purchased by the customer
    # uncommon = [product for product in customer_purchases if product_popularity.get(product, 0) < threshold]

    return uncommon
```

# Cross-Sell

- Data should be in form:
  - Transaction 1: Bread, Milk, Butter
  - Transaction 2: Bread, Diaper, Beer
  - Transaction 3: Milk, Diaper, Beer, Butter
  - Transaction 4: Bread, Milk, Diaper, Beer  
- Duplicate Transactions shouldn't exist (Example : Milk,Milk)  

#### Common Terms:
- Support: The proportion of transactions containing a particular itemset.
- Confidence: The likelihood of purchasing Y given X.
- Lift: The strength of the association between X and Y compared to their individual occurrences.

#### Building the FP-Tree
- Count the frequency of all items
- Remove low frequency items below the support threshold
- Order items in descending frequency
- Construct the tree

```python
# Group by customer and timestamp
# Grouping by customer_id and purchased_date_time to distinguish transactions
basket = df_sales.groupby(['customer_id', 'purchased_date_time'])['product_id'].apply(list).reset_index()
```

```python
basket.head()
```

```python
# Only Taking the required columns
df = pd.DataFrame(basket["product_id"])
```

```python
df.head()
```

```python
# Removing duplicates within the row list
def remove_duplicates(items):
    return list(set(items))  # Remove duplicates by converting to set and back to list

df["product_id"] = df["product_id"].apply(remove_duplicates)
```

```python
df.head()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.fpm import FPGrowth
from pyspark.sql.functions import col
```

```python
# Initialize Spark session
spark = SparkSession.builder \
    .appName("Cross-Sell Analysis with FPGrowth") \
    .getOrCreate()
```

```python
# Create a spark Dataframe from pandas df
df_spark = spark.createDataFrame(df)
```

```python
df_spark.count()
```

```python
# Initialize FP-growth
# We have 73688 transactions
# Setting minimum support to 0.001 indicates that the products must be present in 73688*0.00075 transactions i.e 55 transactions
fp_growth = FPGrowth(itemsCol="product_id", minSupport=0.00075, minConfidence=0.6)
```

```python
# Fit the model
model = fp_growth.fit(df_spark)
```

```python
# Import functions for sorting
from pyspark.sql.functions import size, col

sorted_freq_itemsets = model.freqItemsets.withColumn("item_count", size(col("items"))) \
                                         .orderBy(col("item_count").desc(), col("freq").desc())

sorted_freq_itemsets.show(truncate=False)
```

```python
# Generate association rules
association_rules = model.associationRules
association_rules.show(truncate=False)
```

```python
association_rules_df = association_rules.toPandas()
```

```python
def get_upsell_items(product_id):
    cosine_similarity_series = cosine_similarity_df.loc[product_id]
    ordered_similarities = cosine_similarity_series.sort_values(ascending=False)
    results = ordered_similarities[1:6].index
    product_names = []
    for product in results:
        product_names.append(get_product_name(product))
    return product_names
```

# Top 25 similar products

```python
# Top 25 similarities based on search term
def find_similar_products(search_term,df,top_n = 25):
    descriptions = df['product_description'].tolist()
    descriptions.insert(0, search_term)

    # Convert the text data into numerical vectors using TF-IDF
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform(descriptions)

    # Compute cosine similarity between the search term and all product descriptions
    cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])

    # Add cosine similarity scores to the DataFrame
    df['similarity_score'] = cosine_sim.flatten()

    # Sort products by similarity score (in descending order) and return top N
    similar_products = df.sort_values(by='similarity_score', ascending=False).head(top_n)

    return similar_products[['product_name', 'similarity_score']]
```

```python
from spellchecker import SpellChecker
def autocorrect_sentence(sentence):
    spell = SpellChecker()
    # Split the sentence into words
    words = sentence.split()
    # Correct each word
    corrected_words = [spell.correction(word) for word in words]
    # Join corrected words back into a sentence
    corrected_sentence = " ".join(corrected_words)
    return corrected_sentence
```
